#### MLR

```python
X_mlr_0 = sunflower_dataset[feature_names_0]
y_mlr_0 = sunflower_dataset["VRYIELDMAS"]

scaler_mlr_0 = StandardScaler()
X_mlr_2 = scaler_mlr_0.fit_transform(X_mlr_0)

X_train_mlr_0, X_test_mlr_0, y_train_mlr_0, y_test_mlr_0 = train_test_split(X_mlr_0, y_mlr_0, test_size=0.2, random_state=0)
```
