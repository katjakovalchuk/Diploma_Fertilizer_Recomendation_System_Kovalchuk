### weather

```python
weather_data = pd.read_csv("./data_1/wether_.csv")
```

```python
patterns = 'max_t|min_t|max_h|min_h|max_p|min_p|max_c|min_c'

df = weather_data.drop(weather_data.filter(regex=patterns).columns, axis=1)
df
```

```python
df["location"] = "Ternopil,Uk"
```

```python
df
```

### guideline

```python
guideline = pd.read_csv("./data_1/Fertilizers_data/guideline.csv")
```

```python
guideline.columns
```

```python
# guideline_ = guideline.drop(columns=["Код кольору"])
guideline_ = guideline.rename(
    columns={
        'Довга назва елементу' : "long_elem_name", 
        "Коротка назва елементу" : "short_elem_name",
        'Рівень вмісту' : "content_level",
        'Мінімальне значення' : "min_elem_value", 
        'Максимальне значення' : "max_elem_value",
        "Код кольору" : "color"
    }
)

guideline_["content_level"] = guideline_["content_level"].replace(
    {
        "Дуже низький" : "Very low",
        "Низький" : " Low",
        "Середній" : "Medium", 
        "Високий" : "High", 
        "Дуже високий" : "Very high"
    }
)

guideline_
```

```python
engine = create_engine(
    "postgresql://postgres:postgress@127.0.0.1:5432/postgres"
)
```

```python
guideline_.to_sql("elements_guideline", engine)
```

### fertilizers 

```python
fertlizers = pd.read_csv("./data_1/Fertilizers_data/Fertilizers_table.csv")
```

```python
fertlizers.columns
```

```python
fertlizers_ = fertlizers.rename(
    columns={
        'Назва продукту' : "product_name", 
        "Серія" : "product_line",
        'Тип внесення' : "application_method",
    }
)

fertlizers_["application_method"] = fertlizers_["application_method"].replace(
    {
        "Ґрунтове" : "Soil",
        "Ґрунтове / захищений ґрунт" : " Soil / protected soil",
        "Позакореневе (г/л)" : "Foliar (g/L)", 
        "Обробка насіння (г/л)" : "Seed treatment (g/L)", 
        "Водорозчинне / Краплинне" : "Water-soluble / Drip irrigation",
        "Водорозчинне" : "Water-soluble",
        "Краплинне / Фертигація" : "Drip irrigation / Fertigation",
    }
)

fertlizers_
```

```python
fertlizers_.to_sql("fertilizers_table", engine)
```
